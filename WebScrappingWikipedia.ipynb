```python
import nltk
import urllib
import bs4 as bs
import re
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
#Getting the data source
source = urllib.request.urlopen("https://fr.wikipedia.org/wiki/Science_des_donn%C3%A9es")
```

```python
#Parsing the data / creating BeautifulSoup object
soup = bs.BeautifulSoup(source, 'lxml')
```

```python
#Fetching the data
text = ""
for paragraph in soup.find_all('p'):
    text += paragraph.text
```

```python
text
```

```python
#Preprocessing the scrapped Data using the regular expression pattern "re":
text = re.sub(r'\[[0-9]*\]',' ', text) #Remove [Digits]
text = re.sub(r'\s+',' ', text)   #Remove enpty spaces
text = text.lower()               #Text to lower case 
text = re.sub(r'\d',' ', text)    #Remove digits
text = re.sub(r'\s+',' ', text)   #Remove enpty spaces
```

```python
#Split Datasets to sentences:
sentences =  nltk.sent_tokenize(text)
```

```python
sentences
```

```python
#Split  Sentences to words:
words = [nltk.word_tokenize(sentence) for sentence in sentences]
```
